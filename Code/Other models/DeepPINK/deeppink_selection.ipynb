```python
import numpy as np 
import pandas as pd
```

## DeepPINK simuation

```python
data_s = pd.read_csv('/kaggle/input/deeppink-simulation/result_epoch100_featImport.csv',header=None)
data_c = data_s.mean()
# row_1 = data_c.values
W = []
for i in range(int(0.5 *len(data_c))):
    W.append(data_c[i]**2-data_c [i+100]**2)
```

```python
T = []
for i in W:
    count1 = 0
    count2 = 0
    t = abs(i)
    for j in W:
        if j < -t:
            count1 = count1+1
        if j >= t:
            count2 = count2 +1
    q = (count1)/(max(count2,1))
#     print(count1,count2,q)
    if q <= 0.01:
        T.append(t)
threshold = min(T)
selected_vars = [i+1 for i in range(len(W)) if W[i] >= threshold]
selected_vars
```

## DeepPINK application

```python
data_s = pd.read_csv('/kaggle/input/deeppink-breast-cancer/result_epoch100_featImport.csv',header=None)
data_c = data_s.mean()
# row_1 = data_c.values
W = []
for i in range(int(0.5 *len(data_c))):
    W.append(data_c[i]**2-data_c [i+30]**2)
T = []
for i in W:
    count1 = 0
    count2 = 0
    t = abs(i)
    for j in W:
        if j < -t:
            count1 = count1+1
        if j >= t:
            count2 = count2 +1
    q = (count1)/(max(count2,1))
#     print(count1,count2,q)
    if q <= 0.5:
        T.append(t)
threshold = min(T)
selected_vars = [i+1 for i in range(len(W)) if W[i] >= threshold]
selected_vars
```
